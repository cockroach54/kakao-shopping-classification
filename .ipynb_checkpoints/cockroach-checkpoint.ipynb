```python
y_vocab = _pickle.load(open('./data/y_vocab.py3.cPickle', 'rb'))
```

```python
y_vocab
```

```python
y_vocab['43>109>1576>-1']
```

```python
token_to_cate = {}
for it in y_vocab.items():
     token_to_cate[it[1]] = it[0]
```

```python
token_to_cate
```

```python
dev = [10,334,2029,192]
pid = ['a','b','c','d']
predicted = []
for i in dev:
    tmp = token_to_cate[i]
    predicted.append(tmp.replace('>', '\t'))
```

```python
predicted
```

```python
with open("records.tsv", "w") as f:
    for el in zip(pid, predicted):
#         print(el)
        tmp = '\t'.join(el)      
        f.write(tmp+'\n')
```
